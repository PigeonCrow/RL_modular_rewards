# Dopamine in a world with multiple simultaneous objectives
### Are there multiple pathways for dopamine, corresponding to the task at hand?
<p align="right">presented byDirks, Duarte, Iribarren, Mathew, Voss</p>

Papers discussed:
- Dopaminergic error signals retune to social feedback during courtship - [(Roeser, 2023)](https://doi.org/10.1038/s41586-023-06580-w)
- A feature-specific prediction error model explains dopaminergic heterogeneity - [(Lee, 2024)](https://doi.org/10.1038/s41593-024-01689-1)
- Reward-Bases: Dopaminergic Mechanisms for Adaptive Acquisition of Multiple Reward Types - [(Millidge, 2024)](https://doi.org/10.1371/journal.pcbi.1012580)
- Having multiple selves helps learning agents explore and adapt complex changing worlds - [(Dulberg, 2024)]( https://doi.org/10.1073/pnas.2221180120)

- __Roeser__
    - Biological experiment with zebra finches, showcasing different dopamine pathways in Area X and MST. Dopamine spikes show up differently based on the action that the finch was taking and the social context.
- __Lee__
    - Feature specific prediction errors that showcase how different actions can be chosen by even by the same dopamine in the same dopamine pathways.
- __Millidge__
    - RL model using multiple agents that showcases how the best action is chosen based on the motivation and total reward reward_function.
- __Dulberg__
    - Benefits of modular reward computation against monolithic reward computation.
